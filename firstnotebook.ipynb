```python
import numpy as np #access pi from a python library

def Vol(r):
    volume = 4/3 * np.pi * r**3
    return volume
```

```python
Vol(3.2)
```

```python
k = 1.3
Vol(k)
```

```python
for i in range(10):
    v=Vol(i+1)
    print(i+1, v)
```
